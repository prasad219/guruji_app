```python
flrimg = img.transpose(Image.FLIP_LEFT_RIGHT)
flrimg
```

```python
ftbimg = img.transpose(Image.FLIP_TOP_BOTTOM)
ftbimg
```

```python
rsimg = img.resize((300, 300))
rsimg
```

```python
#4. Write a python program to read a url and do the following:
#● Get the title of the webpage.
#● Get the body of the webpage and create a text file with it.
#● Get all the urls that the web page refers to.
```

```python
import requests
from bs4 import BeautifulSoup


url = "https://www.python.org/"  
res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')
title = soup.title.string
print("Title:",title)
```

```python
body_tag=soup.body

txt=body_tag.text
with open("example1.txt", "w", encoding="utf-8") as f:
    f.write(txt)
f2=open('example1.txt','r')
print(f2.read())
```

```python
soup.find_all('a',href=True)
```

```python
#5. Make an html page and traverse over the tree iteratively, recursively and sideways.
```

```python
pg=open('sample.html','r')
soup = BeautifulSoup(pg, 'html.parser')
h_tag = soup.head
for child in h_tag.children:
    print(child)
```

```python
for sibling in soup.head.next_siblings:
    print(sibling)
```

```python
for sibling in soup.body.previous_siblings:
    print(sibling)
```

```python
for element in soup.body.next_elements:
    print(element)
```

```python
for element1 in soup.head.previous_elements:
    print(element1)
```

```python
#8. Write a python program using scipy to read and write a wavfile.
```

```python
from scipy.io import wavfile

sample_rate, audio_data = wavfile.read('sample.wav')
print("sample rate of sample.wav:",sample_rate)
print("audio data of sample.wav:",audio_data)
wavfile.write('sample2.wav', sample_rate, audio_data)
sample_rate1, audio_data1 = wavfile.read('sample2.wav')
print("sample rate of sample2.wav:",sample_rate1)
print("audio data of sample.wav:",audio_data1)
```
